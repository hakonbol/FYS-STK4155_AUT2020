```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
def MSE(y,ypred):
    # Input y and ypred must be scalars, datapoints of both the true value, y, and the predicted value ypred
    MSE = np.mean((y-ypred**2))
    return MSE

def R2(y,ypred):
    return 1-np.sum((y-ypred)**2)/np.sum((y-np.mean(y))**2)
```

```python
x = np.random.rand(100,1)
y = 2.0+5*x*x+0.1*np.random.randn(100,1)
```

```python
plt.plot(y)
```

```python
# Setting up the design matrix
X = np.zeros((len(x),3))
print(x.shape, X.shape)
X[:,[0]] = 1
X[:,[1]] = x
X[:,[2]] = x**2

#Finding the values for our coefficients with
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
ypred = X @ beta
plt.plot(ypred)
plt.plot(y)
```

```python
# Splitting the data using SciKit Learn
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
beta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
print(beta)
```

```python
ypred_train = X_train @ beta
print("R2 Training:", R2(y_train, ypred_train))
print("MSE Training:", MSE(y_train, ypred_train))
```

```python
# Applying model on the test data and running test scores
ypred_test = X_test @ beta
print("R2 Test:", R2(y_test, ypred_test))
print("MSE Test:", R2(y_test, ypred_test))
```
